```python
df = pd.read_excel('dnd-dataframe.xlsx', sheet_name=0, usecols=['name', 'background'])
df.head(30)
```

```python
df.dropna(inplace=True)
```

```python
blanks = []

for i,lb,rv in df.itertuples():
    if type(rv) == str:
        if rv.isspace():
            blanks.append(i)
```

```python
blanks
```

```python
df.iloc[0]['background']
```

```python
sid.polarity_scores(df.iloc[0]['background'])
```

```python
df['score'] = df['background'].apply(lambda background: sid.polarity_scores(background))
```

```python
df.head()
```

```python
df['compound'] = df['score'].apply(lambda d:d['compound'])
```

```python
df.head()
```

```python
df['comp_score'] = df['compound'].apply(lambda score: 'pos' if score>=0 else 'neg')
```

```python
df.head(30)
```

```python

# First grouping based on "Topic Label" 
# Within each team we are grouping based on "Name" 
gkk = df.groupby(['comp_score', 'compound', 'name'])
  
# Print the first value in each group 
gkk.first() 
```
