```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dnd-dataframe.csv")
data.head(14)
```

```python
dummies = data['weapons'].str.get_dummies(sep=',') 
```

```python

dum_weapon = pd.concat([data, dummies], axis=1) 
```

```python
dum_weapon.head()
```

```python
dummies1 = data['spells'].str.get_dummies(sep=',') 
```

```python
dum_spells = pd.concat([data, dummies1], axis=1) 
```

```python
dum_weapon.head()
```

```python
dum_spells.head()
```

```python
stacked = dum_spells.stack()
index = stacked.index.get_level_values(0)
result = pd.crosstab(index=index, columns=stacked)
result.index.name = None
result.columns.name = None
```

```python
# use pd.concat to join the new columns with your original dataframe
dum_spells = pd.concat([data,pd.get_dummies(data['spells'], prefix='spells')],axis=1)

# now drop the original 'country' column (you don't need it anymore)
dum_spells.drop(['spells'],axis=1, inplace=True)
```

```python
dum_spells.head(13)
```

```python
# use pd.concat to join the new columns with your original dataframe
dum_weapons = pd.concat([dum_spells,pd.get_dummies(dum_spells['weapons'], prefix='weapons')],axis=1)

# now drop the original 'country' column (you don't need it anymore)
dum_weapons.drop(['weapons'],axis=1, inplace=True)
```
